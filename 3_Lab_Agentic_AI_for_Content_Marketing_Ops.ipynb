# Business Use Case: Agentic AI for Content Marketing Ops (Autogen)

## 🎯 Objective  
Automate a content-marketing pipeline—**research → blog drafting → social post → editorial QA**—using **Microsoft Autogen** multi-agent collaboration to increase speed, consistency, and quality.

---

## 🏢 Business Context  
Marketing teams need high-quality, SEO-ready content at pace. Manual workflows are slow, error-prone, and hard to scale.  
This setup uses **Autogen** to orchestrate specialized agents that collaborate to produce research-grounded, publish-ready assets (blog + tweet) for any topic.

---

## 🤖 Agents Involved

1. **Researcher (Online Researcher)**  
   - **Function:** Perform open-web research; collect factual notes and reputable links.  

  >  https://www.tavily.com/

  >  https://app.tavily.com/home
   - **Tools:** **Tavily web search** + custom `process_search(url)` scraper (BeautifulSoup).  
   - **Business Value:** Evidence-backed inputs reduce hallucinations and rework.

2. **Blog Manager (SEO Writer)**  
   - **Function:** Draft a multi-paragraph, **SEO-optimized** blog with **inline citations/links**.  
   - **Business Value:** Faster drafting with consistent structure and on-brand tone.

3. **Social Media Manager (Comms)**  
   - **Function:** Distill the blog’s key message into a **≤280-char, high-engagement tweet**.  
   - **Business Value:** Omnichannel coverage with minimal lag from longform to shortform.

4. **Content Marketing Manager (Editor/QA)**  
   - **Function:** Review for tone, accuracy, SEO elements, and policy/safety; return **APPROVED** or fix-list.  
   - **Business Value:** Prevents brand, legal, and quality issues before publication.

*(Autogen Orchestration: `GroupChat` + `GroupChatManager` + `UserProxyAgent` for hands-off runs.)*

---

## 🧩 Tasks Executed (Pipeline)

1. **Gather Source Material**  
   - **Inputs:** Topic query (e.g., “Agentic Behavior”).  
   - **Actions:** Tavily search → follow links → `process_search(url)` extracts **visible text** for notes.  
   - **Output:** Bullet-point research summary + source links.

2. **Draft SEO Blog**  
   - **Inputs:** Research notes + links.  
   - **Actions:** Compose 3+ paragraph article, headings, keywords, inline citations.  
   - **Output:** Source-linked, SEO-aware blog draft.

3. **Create Social Summary**  
   - **Inputs:** Approved blog content.  
   - **Actions:** Craft compelling tweet (≤280 chars), optional contextual hashtag(s).  
   - **Output:** Tweet copy with link hook.

4. **Editorial Review & Safety Check**  
   - **Inputs:** Blog + tweet.  
   - **Actions:** Check cohesion, accuracy, SEO, policy/safety; return APPROVED or corrections.  
   - **Output:** Final sign-off or revision guidance.

---

## 🔁 Orchestration & Data Flow  
- **Engine:** Autogen `GroupChat` coordinates specialized agents.  
- **Intermediate Artifacts:** Shared via the group chat context (notes → blog → tweet → review).  
- **Tool Use:** Agents can call `process_search(url)` as an OpenAI function to ingest page text.  
- **LLM Config:** Centralized `llm_config` (model, temperature, tool schemas).

---

## 📈 Business Benefits  
- **Speed:** Research-to-publish cycle time reduced dramatically.  
- **Quality:** Evidence-grounded drafts with built-in editorial QA.  
- **Consistency:** Standardized voice, SEO, and safety checks.  
- **Scalability:** Spin up concurrent topics with identical workflow logic.

## 🛠️ Tech Notes

---

## ⚠️ Risks & Mitigations  
- **Outdated/biased sources:** Restrict to reputable domains; timestamp checks; require ≥2 corroborating sources.  
- **Policy violations:** Final gate via Editor agent with explicit safety rubric.  
- **SEO underperformance:** Add SEO Analyst agent; A/B test titles, intros, and CTAs.  
- **Key leakage:** Ensure secrets via env/secret store; zero logging of creds.

```python




def process_search(url: str) -> str:
    """Fetches *visible* text from the provided URL."""
    res = requests.get(url, timeout=15)
    soup = BeautifulSoup(res.content, "html.parser")
    return soup.get_text(" ", strip=True)

# Tool schemas for OpenAI function calling
process_search_schema = {
    "name": "process_search",
    "description": "Fetches *visible* text from the provided URL.",
    "parameters": {
        "type": "object",
        "properties": {
            "url": {"type": "string", "description": "The URL to fetch and extract text from."}
        },
        "required": ["url"]
    }
}

# Use autogen's OpenAI config format with functions for agents
llm_config = {
    "model": "gpt-3.5-turbo",
    "api_key": os.environ["OPENAI_API_KEY"],
    "temperature": 0.4,
    "functions": [process_search_schema],
}
# Minimal config for GroupChatManager (no functions/tools)
groupchat_llm_config = {
    "model": "gpt-4o",
    "api_key": os.environ["OPENAI_API_KEY"],
    "temperature": 0.4,
}
```

```python

# ───────────────────────────────────────────────────────────────────────────────
# Agent definitions
# ───────────────────────────────────────────────────────────────────────────────

def make_agent(name: str, role: str, goal: str, backstory: str):
    agent = AssistantAgent(
        name=name,
        llm_config=llm_config,
        system_message=f"You are {name}, acting as a {role}. Goal: {goal}. {backstory}",
    )
    return agent

# Create the agents
researcher = make_agent(
    "researcher",
    "Online Researcher",
    "Compile factual notes on a query",
    "You excel at open‑web exploration, discerning reputable sources, and producing concise fact sheets.",
)
researcher.register_function({"process_search": process_search})

blog_manager = make_agent(
    "blog_manager",
    "Blog Manager",
    "Draft SEO‑optimised, source‑linked blog posts",
    "Your drafts are engaging, structured, and align with a target audience's reading level.",
)
blog_manager.register_function({"process_search": process_search})

social_manager = make_agent(
    "social_manager",
    "Social Media Manager",
    "Compress ideas into a compelling 280‑character tweet",
    "You craft concise, high‑engagement tweets using hashtags when helpful.",
)
social_manager.register_function({"process_search": process_search})

content_manager = make_agent(
    "content_manager",
    "Content Marketing Manager",
    "Review and approve deliverables, flagging policy violations",
    "You check tone, accuracy, SEO elements, and ensure no hateful or explicit language appears.",
)
content_manager.register_function({"process_search": process_search})

# Create a UserProxyAgent to orchestrate the workflow
user_proxy = UserProxyAgent(
    name="user_proxy",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=10,
    code_execution_config={"use_docker": False},
)

def main():
    # Define the group chat with all agents
    groupchat = GroupChat(
        agents=[user_proxy, researcher, blog_manager, social_manager, content_manager],
        messages=[],
        max_round=20,
    )
    manager = GroupChatManager(
        groupchat=groupchat,
        llm_config=groupchat_llm_config,
    )
    # Start the workflow
    user_proxy.initiate_chat(
        manager,
        message=(
            "Research the topic 'Agentic Behavior'. Provide bullet-point notes plus links. "
            "Then, using the research, draft a 3+ paragraph blog post with inline links. "
            "Summarize the key takeaway from the blog post as a single tweet (≤280 chars). "
            "Finally, review the blog post and tweet for cohesion, safety, and SEO. Respond with 'APPROVED' or corrections."
        ),
    )

if __name__ == "__main__":
    main()
```
